### Year with high suicide rate

```python
temp=df.groupby('year')['suicides_no'].sum().reset_index().sort_values('suicides_no',ascending=False)
temp
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='year',y='suicides_no',data=temp);
```

### Top 10 countries with high suicide rate

```python
temp=df.groupby('country')['suicides_no'].sum().reset_index()
temp=temp.sort_values('suicides_no',ascending=False)
```

```python
plt.figure(figsize=(15,8))
ax=sns.barplot(x='country',y='suicides_no',data=temp.head(10));
plt.xlabel('Countries')
plt.xlabel('Number of Suicides')
plt.title('Top 10 countries with high suicide rate')
```

### Gender

```python
df['sex'].value_counts()
```

###  Suicide rate in male and female

```python
temp=df.groupby('sex')['suicides_no'].sum().reset_index()
temp
```

```python
sns.barplot(x='sex',y='suicides_no',data=temp);
```

```python
temp=df.groupby(['country','sex'])['suicides_no'].sum().reset_index()
temp=temp.sort_values('suicides_no',ascending=False)
temp
```

#### top 10 countries with High suicide rate in male

```python
plt.figure(figsize=(15,5))
sns.barplot(x='country',y='suicides_no',data=temp[temp['sex']=='male'].head(10));
```

#### top 10 countries with High suicide rate in female

```python
plt.figure(figsize=(15,5))
sns.barplot(x='country',y='suicides_no',data=temp[temp['sex']=='female'].head(10));
```

### Age Category

```python
temp=df.groupby('age')['suicides_no'].sum().reset_index()
temp
```

#### Highest Suicide rate in age category

```python
sns.barplot(x='age',y='suicides_no',data=temp,order=['5-14','15-24','25-34','35-54','55-74','75+']);
```
